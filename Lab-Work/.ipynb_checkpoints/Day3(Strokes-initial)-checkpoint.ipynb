## 1.1 changing data types

- let's change the id format from integer to str (object)
- let's change data type from integer to boolean

```python
strokes['id'] = strokes['id'].astype(str)
```

```python
strokes[['heart_disease', 'hypertension']] = strokes[['heart_disease', 'hypertension']].astype(bool)
```

```python
strokes.info()
```

```python
# python dictionaries

#[] list
#() in pandas - bunch of variable you can use! 
#{} dictionary
#key + value

#key = sian
#value = 38
#value = red trousers
#value = spain

#No: false
#Yes: true
    
strokes = strokes.replace({'No':False, 'Yes': True})
```

```python
strokes.info()
```

#### create 3 data frames
- df1 = rows 1- 1000 index  :1000   exclusive on top end, inclusive on bottom end
- df2 = rows 1001- 3500 index  1000:3500
- df3 = the rest index 3500:


```python
strokes.head(2) # index always starts at zero
```

```python
#iloc # index location
df1 = strokes.iloc[:1000]
df2 = strokes.iloc[1000:3500]
df3 = strokes.iloc[3500:]
```

```python
df1.info()
```

```python
df2.info()
```

```python
df2.head() # reset index on df2 and df3
```

```python
df2.reset_index(drop = True, inplace = True)
df3.reset_index(drop = True, inplace = True)
```

```python
df2.info()
```

```python
df3.info()
```

```python
df_new = pd.concat([df1,df2,df3], axis = 0)
```
